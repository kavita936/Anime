```python
def extract_episode(txt):
  check = False # Corrected: 'false' should be 'False'
  data = ""
  for i in txt:
    if i == '(': # Start of potential episode data
      check = True # Corrected: 'true' should be 'True'
    elif i == ')': # End of episode data
      break # Exit the loop once the closing parenthesis is found
    if check and i != '(': # If capturing and not the opening parenthesis itself
      data += i # Corrected: `data += 1` would cause a TypeError. Assuming intent is to append character.
  return data # Return the extracted data
```

```python
df1['episodes'] = df1["Title"].apply(extract_episode)
```

```python
df1['episodes'] = df1['episodes'].astype(str).str.replace(r'\D', '', regex=True).astype(int)
```

```python
def extraction_time(txt):
  data = ""
  for i in range(len(txt)): # Corrected: Iterate by index
    if txt[i] == ')': # Corrected: Check character at index 'i' and add colon
      start_extract = i + 1
      # Extract 18 characters starting from after the ')'
      end_extract = min(start_extract + 18, len(txt)) # Ensure not to go out of bounds
      data = txt[start_extract:end_extract].strip() # Extract and strip whitespace
      return data # Return immediately after finding the first ')'
  return data # If no ')' is found, return empty string
```

```python
df1['total time'] = df1['Title'].apply(extraction_time)
```
